```python
class Solution(object):
    def removeElement(self, nums, val):
        k = 0
        for i in range(len(nums)):
            if nums[i] != nums[k]:
                nums[k] = nums[i]
                k+=1
        return k
```

```python
arr = [1,2,2,2,3,4,5,6]
val = 2
obj = Solution()
obj.removeElement(arr, val)
```
